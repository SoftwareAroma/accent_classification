```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load the audio files
a,s=librosa.load("../../data/audio/arabic1.wav")
b,t=librosa.load("../../data/audio/english1.wav")
```

```python
# get the mfcc features
mfcc1= librosa.feature.mfcc(y=a,sr=s)
mfcc2= librosa.feature.mfcc(y=b,sr=t)
```

```python
# concet the features
def concatfeatures(realdata, mfcc1):
    temp1=mfcc1.T[1][1:19]
    for frame in range(10,50):
        temp1=np.concatenate((temp1,mfcc1.T[frame][1:19]))
    #print(temp1)
    realdata.append(temp1)
```

```python
real=[]
concatfeatures(real,mfcc1)
concatfeatures(real,mfcc2)
```

```python
plt.plot(real[0])
plt.plot(real[1])
# plt.plot(real[2])
# plt.plot(real[3])
plt.xlabel("Concatenated MFCC Features")
plt.ylabel("Amplitude")
plt.legend(["Arabic","English"])
plt.savefig("../../output/analysis.png")
plt.show()
```
